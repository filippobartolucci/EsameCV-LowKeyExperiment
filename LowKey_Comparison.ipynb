# LowKey Results Comparison

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
import os
```

```python
imgs_dir = "./LowKey_code/imgs"
imgs_files = os.listdir(imgs_dir)

for img in imgs_files:
    if "attacked" in img:
        continue

    img_path = os.path.join(imgs_dir, img)
    img = cv2.imread(img_path)

    attached_path = img_path[:-4] + "_attacked.png"
    attacked_img = cv2.imread(attached_path)

    fig, ax = plt.subplots(1, 3, figsize=(15, 15))
    
    for i in range(3):
        ax[i].set_axis_off()

    ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax[0].set_title("Original")
    ax[1].imshow(cv2.cvtColor(attacked_img, cv2.COLOR_BGR2RGB))
    ax[1].set_title("Attacked")

    difference = cv2.absdiff(img, attacked_img)
    ax[2].imshow(difference)
    plt.show()

    
```
