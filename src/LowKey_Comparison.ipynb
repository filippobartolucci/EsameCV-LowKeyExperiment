```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
import os
from util.prepare_utils import prepare_models, prepare_dir_vec, get_ensemble, prepare_data, extract_features
import torchvision.transforms as transforms
import my_utils
```

# Comparing original and attacked images

```python
imgs_dir = "./imgs"
imgs_files = os.listdir(imgs_dir)

for img in imgs_files:
	if "attacked" in img or "small" in img or "face" in img:
		continue

	img_path = os.path.join(imgs_dir, img)
	img = cv2.imread(img_path)

	face_path= img_path[:-4] + "_face.png"
	face = cv2.imread(face_path)
	attached_path = face_path[:-4] + "_attacked.png"
	attacked_img = cv2.imread(attached_path)

	fig, ax = plt.subplots(1, 3, figsize=(15, 15))
	
	for i in range(3):
		ax[i].set_axis_off()

	ax[0].imshow(cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
	ax[0].set_title("Original")
	ax[1].imshow(cv2.cvtColor(attacked_img, cv2.COLOR_BGR2RGB))
	ax[1].set_title("Attacked")

	difference = cv2.absdiff(face, attacked_img)
	ax[2].imshow(difference)

	plt.show()
```
